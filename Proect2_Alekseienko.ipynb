Завдання 1. Підрахунок загальної кількості респондентів

```python
# Підрахунок кількості унікальних респондентів за їхнім ID
total_respondents = df['ResponseId'].nunique()

# Виведення результату
total_respondents
```

Завдання 2. Аналіз повноти відповідей респондентів

```python
#  Отримуємо унікальні назви питань зі схеми (поле qname)
schema_questions = set(df1['qname'].unique())

#  Отримуємо назви колонок, які є в основному датасеті
data_columns = set(df.columns)

#  Знаходимо перетин (спільні колонки, які є і там, і там)
common_columns = list(schema_questions.intersection(data_columns))

# Фільтруємо основний датасет, залишаючи лише ці колонки
df_filtered = df[common_columns]

#  Рахуємо кількість респондентів, у яких немає жодного пропуску (NaN) у цих колонках
# Метод .dropna() видаляє рядки з пропусками, .shape[0] поверне кількість рядків, що залишилися
respondents_full_answers = df_filtered.dropna().shape[0]

# Виведення результату
print(f"Кількість респондентів, які відповіли на всі вибрані питання: {respondents_full_answers}")
```

Завдання 3. Статистичний аналіз досвіду респондентів

```python
# Розрахунок середнього значення (mean)
mean_exp = df['WorkExp'].mean().round(1)

# Розрахунок медіани (median)
median_exp = df['WorkExp'].median()

# Розрахунок моди (mode) - беремо перший елемент [0], якщо значень декілька
mode_exp = df['WorkExp'].mode()[0]

# Формування підсумкової таблиці (Series) для зручного відображення
central_tendency = pd.Series({
    'Середнє (Mean)': mean_exp,
    'Медіана (Median)': median_exp,
    'Мода (Mode)': mode_exp
})

# Виведення результату
central_tendency
```

Завдання 4. Аналіз віддаленої роботи

```python
# Знаходимо кількість респондентів зі значенням 'Remote'
remote_workers_count = df[df['RemoteWork'] == 'Remote'].shape[0]

# Виведення результату
remote_workers_count
```

Завдання 5. Визначення популярності Python

```python
#  Створюємо новий стовпець (True, якщо вказано Python, інакше False)
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)

# Рахуємо кількість тих, хто вказав Python
python_users_count = df['worked_with_python'].sum()

# Рахуємо загальну кількість респондентів через унікальні ID
total_respondents = df['ResponseId'].nunique()

# Обчислюємо відсоток від УСІХ учасників
python_percentage = (python_users_count / total_respondents) * 100

# Виведення результату
print(f"{python_percentage.round(2)}%")

```

Завдання 6. Аналіз шляхів навчання програмуванню

```python
# Рахуємо кількість респондентів, у яких в списку навчання є 'Online Courses'
# na=False дозволяє пропустити тих, хто не відповів на це питання
online_learners_count = df['LearnCode'].str.contains('Online Courses', na=False).sum()

# 2. Виведення результату
online_learners_count
```

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
#  Фільтруємо лише Python-розробників та прибираємо тих, хто не вказав зарплату
python_devs = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)].copy()

# Групуємо за країною та обчислюємо середнє (mean) і медіану (median) для річної компенсації
# Pandas автоматично ігнорує пропуски (NaN) у ConvertedCompYearly при розрахунках
country_stats = python_devs.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).dropna()

# Перейменовуємо колонки для зручності
country_stats.columns = ['Середня компенсація', 'Медіанна компенсація']

# Округляємо результат до 2 знаків після коми та виводимо перші 15 країн
country_stats.round(2).head(15)

```

```python
# Створюємо кольорову таблицю
colored_table = country_stats.head(15).style.background_gradient(cmap='YlGnBu')

# Відображаємо її
colored_table
```

Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

```python
# Фільтруємо пропуски, вибираємо 3 стовпці та сортуємо за спаданням зарплати
top_5_specialists = (
df.dropna(subset=['ConvertedCompYearly'])
                     [['ResponseId', 'EdLevel', 'ConvertedCompYearly']]
                     .sort_values(by='ConvertedCompYearly', ascending=False)
                     .reset_index(drop=True)
                     .head(5))

# Виведення результату
top_5_specialists
```

*Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
# Ознака володіння Python створена у завданні №5 (worked_with_python)
# Групуємо за віком та рахуємо середнє значення (це і є частка True)
age_python_stats = df.groupby('Age')['worked_with_python'].mean() * 100

# Перетворюємо у DataFrame та перейменовуємо колонку
age_python_df = age_python_stats.reset_index()
age_python_df.columns = ['Вікова категорія', 'Відсоток Python-розробників']

# Округляємо та виводимо результат
age_python_df.round(2)
```

*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
# Визначаємо 75-й перцентиль для річної компенсації
income_threshold = df['ConvertedCompYearly'].quantile(0.75)

# Фільтруємо респондентів: зарплата > 75% ТА віддалена робота
high_paid_remote = df[
    (df['ConvertedCompYearly'] >= income_threshold) & 
    (df['RemoteWork'] == 'Remote')
].copy()

# Підраховуємо частоту індустрій (видаляємо пропуски)
industry_stats = high_paid_remote['Industry'].value_counts()

# Перетворюємо у зручний DataFrame для виводу
industry_report = industry_stats.reset_index()
industry_report.columns = ['Індустрія', 'Кількість респондентів']

# Виведення результату (Топ-10 індустрій)
industry_report.head(10)
```

```python
import matplotlib.pyplot as plt

industry_report.head(10).plot(kind='barh', x='Індустрія', y='Кількість респондентів', color='skyblue')
plt.title('Top 10 Industries for High-Paid Remote Workers')
plt.xlabel('Number of Respondents')
plt.gca().invert_yaxis() # щоб Software Development був зверху
plt.show()
```
